```python
#librerias requeridas
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#carga desde archivo .csv sin indice
df=pd.read_csv('Victoria.csv')
df.head()
```

```python
df.info()
```

```python
#identificación de valores nulos 
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
vn=df.isnull().sum().sum()
vn
```

```python
#eliminamos columnas innecesarias 
df2= df.drop(['scrape_id', 'last_scraped', 'source','host_id','host_url',
              'host_location', 'host_about', 'host_acceptance_rate', 'host_thumbnail_url',
              'host_picture_url', 'host_neighbourhood', 'host_listings_count',
              'host_total_listings_count','host_verifications', 'host_has_profile_pic',
              'neighbourhood_group_cleansed', 'minimum_minimum_nights',
              'maximum_minimum_nights', 'minimum_maximum_nights','maximum_maximum_nights','minimum_nights_avg_ntm',
              'maximum_nights_avg_ntm', 'calendar_last_scraped','calendar_updated','license'], axis=1)
df2
```

```python
df2.info()
```

```python
#identificamos valores nulos
vn=df2.isnull().sum()
vn
```

```python
#identificamos valores nulos del dataframe 
vn=df2.isnull().sum().sum()
vn
```

```python
#hacemos una copia de df2
df3=df2.copy()
```

**MÉTODO DE SUSTITUCIÓN POR PROMEDIO**

```python
#reviews_per_month. Para esta variable, debido a los datos que maneja es conveniente utilizar promedio para añadir valores más claros de acuerdo a las 
#tendencias en reseñas que el cliente puede encontrar
df3["reviews_per_month"]=df3["reviews_per_month"].fillna(round(df2["reviews_per_month"].mean(),1))
```

```python
#corroboramos valores nulos 
vn=df3.isnull().sum()
vn
```

```python
#para las variables review_scores_rating, review_scores_accuracy , review_scores_location, review_scores_value,es importante aplicar una sustitución 
#por promedio, debido a que sus datos son necesarios para un análisis adecuado de las necesidades del cliente, necesitamos obtener valores 
#que sean claros

df3["review_scores_rating"]=df3["review_scores_rating"].fillna(round(df2["review_scores_rating"].mean(),1))
```

```python
#review_scores_accuracy
df3["review_scores_accuracy"]=df3["review_scores_accuracy"].fillna(round(df2["review_scores_accuracy"].mean(),1))
```

```python
#review_scores_location
df3["review_scores_location"]=df3["review_scores_location"].fillna(round(df2["review_scores_location"].mean(),1))
```

```python
#review_scores_value
df3["review_scores_value"]=df3["review_scores_value"].fillna(round(df2["review_scores_value"].mean(),1))
```

```python
#price.Para esta variable, debido a los datos que maneja es conveniente utilizar promedio para añadir valores más claros de acuerdo a las 
#tendencias en precios que el cliente desea encontrar
df3['price'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '').str.replace(',', ''), errors='coerce')
df3['price'].fillna(round(df3['price'].mean(), 1), inplace=True)
```

```python
#corroboramos valores nulos 
vn=df3.isnull().sum()
vn
```

**MÉTODO DE SUSTITUCIÓN POR MEDIANA**

```python
#bathrooms. Para esta variable aplicamos método de sustitución por mediana debido a que presenta valores discretos, por ende su sustitución será robusta 
df3["bathrooms"]=df3["bathrooms"].fillna(round(df2["bathrooms"].median(), 1))
```

```python
#corroboramos valores nulos
vn=df3.isnull().sum()
vn
```

```python
#bedrooms. Para esta variable aplicamos método de sustitución por mediana debido a que presenta valores discretos, por ende su sustitución será robusta 
#ya que no se manejan datos tan dispersos. 
df3["bedrooms"]=df3["bedrooms"].fillna(round(df2["bedrooms"].median(), 1))
```

```python
#beds. Al igual que las variables anteriores, al ser valores numéricos pequeños utilizar la mediana es de gran utilidad y obtendríamos valores típicos
df3["beds"]=df3["beds"].fillna(round(df2["beds"].median(), 1))
```

```python
#corroboramos valores nulos 
vn=df3.isnull().sum()
vn
```

**MÉTODO DE SUSTITUCIÓN POR VALOR NUMÉRICO EN CONCRETO**

```python
#para las variables review_scores_cleanliness, review_scores_checkin y review_scores_communication utilizamos valores numéricos exactos dentro del rango
#en que se encuentran sus datos con el fin de que los valores a modificar no presenten un desbalance entre los ya existentes
df3["review_scores_cleanliness"]=df3["review_scores_cleanliness"].fillna(5)
df3["review_scores_checkin"]=df3["review_scores_checkin"].fillna(5)
df3["review_scores_communication"]=df3["review_scores_communication"].fillna(5)
```

```python
#corroboramos valores nulos 
vn=df3.isnull().sum()
vn
```

**MÉTODO DE SUSTITUCIÓN POR UN STRING EN CONCRETO**

```python
#Para la variable descripción, al ser un objeto donde se almacenan cadenas de caracteres mismas que podrían afectar los datos proporcionados es 
#recomendable mantener su uniformidad sustituyendo el valor por un string en concreto descriptivo de la ausencia del dato. 
df3["description"] = df3["description"].fillna("No se proporciona descripción")
```

```python
#corroboramos valores nulos 
vn=df3.isnull().sum()
vn
```

```python
#Para neighborhood_overview utilizamos sustitución por string debido a que los datos presentados no pueden ser alterados para una descripción importante
#indicaremos por ello que no se presenta la información
df3["neighborhood_overview"] = df3["neighborhood_overview"].fillna("No se proporcionan detalles")
```

```python
#corroboramos valores nulos
vn=df3.isnull().sum()
vn
```

```python
#En neighbourhood, podemos sustituir or un string en concreto, debido a que los datos contenidos en esa columna se repiten
#sustituir por un string entonces no afectaría los datos presentados
df3["neighbourhood"] = df3["neighbourhood"].fillna("Neighborhood highlights")
```

```python
#corroboramos valores nulos
vn=df3.isnull().sum()
vn
```

```python
#host_response_time al ser una variable tipo objeto que almacena cadenas de caracteres el método de sustitución por un string en concreto 
#resuta util para indicar que el tiempo de respuesta del host es no fue indicado y no alterar los datos. 
df3["host_response_time"] = df3["host_response_time"].fillna("Desconocido")
```

```python
#corroboramos valores nulos
vn=df3.isnull().sum()
vn
```

```python
#host_response_rate. Al igual que la variable anterior, utilizamos um string para indicar que el tiempo de respuesta del anfitrión no fue
#indicado por el huesped con un -% y de esta manera al obtener resultados la tasa del porcentaje no se vea afectada
df3["host_response_rate"] = df3["host_response_rate"].fillna("-%")
```

```python
#corroboramos valores nulos 
vn=df3.isnull().sum()
vn
```

```python
#first_review. Para esta variable, como se trabaja con fechas utilizare un string que indique la primera revisión con la fecha de cuando se están llenando 
#los datos faltantes, para saber a que datos hay que ponerles mayor atención. 
df3["first_review"] = df3["first_review"].fillna("23/01/2025")
```

```python
#last_review. También es una variable que trabaja con fechas, por ello se utilizará un string en concreto para definir una fecha como ultima revisión 
#y saber que filas han sido modificadas 
df3["last_review"] = df3["last_review"].fillna("24/01/2025")
```

```python
#corroboramos valores nulos 
vn=df3.isnull().sum()
vn
```

**MÉTODO DE SUSTITUCIÓN DE VALORES NULOS POR VALORES NO NULOS HACIA DELANTE O HACIA ATRÁS**

```python
#para la variable host_is_superhost utilizamos bfill debido a que los datos que acepta esta variable son unicamente dos y estos pueden ser 
#remplazados por los valores posteriores
df3["host_is_superhost"] = df3["host_is_superhost"].fillna(method="bfill")
```

```python
#para la variable has_availability, al igual que la anterior, unicamente acepta dos valores, verdadero o falso, por lo tanto puede ser reemplazada por 
#los valores posteriores a ella sin afectar los datos. 
df3["has_availability"] = df3["has_availability"].fillna(method="bfill")
```

```python
#para la variable bathrooms_text podemos aplicar metodo de sustitución hacia atrás debido a que los datos que trabaja se encuentran dentro del mismo
#rango.  
df3["bathrooms_text"] = df3["bathrooms_text"].fillna(method="bfill")
```

```python
#corroboramos valores nulos
vn=df3.isnull().sum()
vn
```

```python
#verificamos nulos del dataframe 
vn=df3.isnull().sum().sum()
vn
```
